## Code Explaination
To Code this, I needed to create a couple simple functions. First I needed a function to translate a list of words to a dictionary of those words with those frequencies. Next I needed a function that took that dictionary and added the relative frequency, or the probability, or each word by dividing the frequency with the total number of words. In this project, these values are only needed for the output. In the future we will need these probabilities to make better guesses. Lastly, I get the users input. If not found in the list of words, I need to suggest the 5 closest words using Levenshtein method of the Minimum Edit Disatnce Algorithm.

```python
import nltk
corpus_name = "brown"           # Change This
nltk.download(corpus_name)
corpus = nltk.corpus.brown      # Change This
```

```python
def word_frequencies(words):
    freq = {}
    for word in words:
        if word in freq:
            freq[word] += 1
        else:
            freq[word] = 1
    return freq
```

```python
N = word_frequencies(corpus.words())
print("Length of corpus (N)", len(N))
print(N)
```

```python
def relative_frequency(freq, total):
    rel_freq = {}
    for word in freq:
        rel_freq[word] = freq[word] / total
    return rel_freq
```

```python
W = relative_frequency(N, len(corpus.words()))
print("Relative frequency of words (W)\n", W)
```

```python
while True:
    user_input = input("Enter something (type ':q' to quit): ")
    if user_input.lower() == ':q':
        print("Goodbye!")
        break
    if user_input in N:
        print(user_input, "is a complete and correct word as per corpus" , corpus_name, ", and its probability is", W[user_input])
    else:
        print(user_input, " not found in corpus.")
        levenshtein_distance = {}
        for word in N:
             levenshtein_distance[word] = nltk.edit_distance(user_input, word)
        
        print("5 most similar words:")
        # print sorted dictionary by value and the words relative frequency using W
        for word in sorted(levenshtein_distance, key=levenshtein_distance.get)[:5]:
            print("\t", word, "with distance", levenshtein_distance[word], "and probability", W[word])
    print("\n=====================================\n")
    
```
